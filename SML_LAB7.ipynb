4.Find MSE, MAE, for test size (20,25)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Sample data
data = {
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5],
    'Salary': [39343.0, 46205.0, 37731.0, 43525.0, 39891.0, 56642.0, 60150.0, 54445.0, 64445.0, 57189.0, 63218.0, 55794.0, 56957.0, 57081.0, 61111.0, 67938.0, 66029.0, 83088.0, 81363.0, 93940.0, 91738.0, 98273.0, 101302.0, 113812.0, 109431.0, 105582.0, 116969.0, 112635.0, 122391.0, 121872.0]
}

df = pd.DataFrame(data)

# Features and target variable
X = df[['YearsExperience']]
y = df['Salary']

# Function to compute MSE and MAE for a given test size
def calculate_metrics(test_size):
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate MSE and MAE
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    return mse, mae

# Test sizes
test_sizes = [20, 25]

for size in test_sizes:
    mse, mae = calculate_metrics(size / len(df))
    print(f"Test Size: {size}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print("-" * 30)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset
data = {
    'Feature 1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature 2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature 3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature 4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target(Y)': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Extract features and target variable
X = df[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']]
y = df['Target(Y)']

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients and intercept
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")

# Predict y values
y_pred = model.predict(X)

# Plot the data points and predictions
plt.figure(figsize=(12, 8))

# Scatter plot for each feature vs Target(Y)
for i, feature in enumerate(X.columns):
    plt.subplot(2, 2, i + 1)
    plt.scatter(df[feature], y, color='blue', label='Data Points')
    plt.plot(df[feature], y_pred, color='red', label='Best Fit Line')
    plt.xlabel(feature)
    plt.ylabel('Target(Y)')
    plt.title(f'{feature} vs Target(Y)')
    plt.legend()

plt.tight_layout()
plt.show()
```

4. Find MSE , MAE for test size (20,30)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Sample data
data = {
    'Feature 1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature 2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature 3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature 4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target(Y)': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Extract features and target variable
X = df[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']]
y = df['Target(Y)']

# Function to compute MSE and MAE for a given test size
def calculate_metrics(test_size):
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate MSE and MAE
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    return mse, mae

# Test sizes
test_sizes = [0.2, 0.3]

for size in test_sizes:
    mse, mae = calculate_metrics(size)
    print(f"Test Size: {int(size * 100)}%")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print("-" * 30)
```
